<div  class="alert alert-block alert-info"style="font-family:cursive;">
Aus der Beschreibung können wir ersehen, dass Date_of_Journey ein Objektdatentyp ist,
Daher müssen wir diesen Datentyp zu Zeit umwandeln, um diese Spalte richtig für die Vorhersage später zu verwenden.<br><br>
Das Modell kann diese String-Werte nicht verstehen, es versteht nur den datetime Form-
Dazu benötigen wir die Funktion to_datetime, um den Objektdatentyp in datetime dtype umzuwandeln.


Die Methode dt.day wird nur den Tag aus  dieses Datums extrahieren
Die Methode dt.month wird nur den Monat aus dieses Datums extrahieren

</div>

```python
# object spalte zu datetime konvertieren
def change_into_datetime(col):
    train_data[col]= pd.to_datetime(train_data[col])
```

```python
#train data columns anschauen
train_data.columns
```

```python
for i in ['Date_of_Journey','Dep_Time', 'Arrival_Time']:
    change_into_datetime(i)
```

```python
# ergebnisse anzeigen
train_data.dtypes
```

```python
# tag und monat extrahieren und in zwei spalten (journey_day,journey_mounth) hinzüfügen
train_data['journey_day']=train_data['Date_of_Journey'].dt.day
train_data['journey_mounth']=train_data['Date_of_Journey'].dt.month
```

```python
train_data.head()
```

```python
# data of journey spalte löschen
train_data.drop('Date_of_Journey',axis=1,inplace=True)
```

<div  class="alert alert-block alert-info"style="font-family:cursive;">
Abflugzeit ist, wenn ein Flugzeug das Gate verlässt.<br>
wie bei Date_of_Journey können wir Werte aus Dep_Time extrahieren

</div>

```python
def extract_hour(df,col):
    df[col+"_hour"]=df[col].dt.hour
def extract_min(df,col):
    df[col+"_minute"]=df[col].dt.minute
def drop_column(df,col):
    df.drop(col,axis=1,inplace=True)
```

```python
#minuten, und stunden aus der spalte 'Dep_time' extrahieren
extract_hour(train_data,'Dep_Time')
extract_min(train_data,'Dep_Time')
# die spalte 'Dep_time' löschen
drop_column(train_data,'Dep_Time')
```

```python
train_data.head()
```

<div  class="alert alert-block alert-info"style="font-family:cursive;">
Ankunftszeit ist, wenn das Flugzeug am Gate ankommt.<br>
wie bei Dep_time können wir Werte aus 'Arrival_time' extrahieren

</div>

```python
# Extracting Hours
extract_hour(train_data,'Arrival_Time')

# Extracting minutes
extract_min(train_data,'Arrival_Time')

# Now we can drop Arrival_Time as it is of no use
drop_column(train_data,'Arrival_Time')
```

## Abgeleitete Merkmale aus Daten extrahieren

```python
## um später unsere model fütter zu können
## ist notwendig paar spalten in deselbe format umzuwandeln ---> später stunden und minuten extrahieren
duration = list(train_data['Duration'])
```

```python
x='2h 50m'
len(x.split(' '))
```

<div  class="alert alert-block alert-info"style="font-family:cursive;">
Die  spalte 'Duration' vorbereiten <br>
-----> 1) alle spalte element in desselbe form umwandeln Z.b '2h 0m'
    <br>
-----> 2) stunde und minute extrahieren Z.B '2h 0m' => 2 - 0 
</div>

```python
# 1/
for i in range(len(duration)):
    if len(duration[i].split(' '))==2:
        pass
    else:
        if 'h' in duration[i]:
            duration[i]=duration[i]+' 0m'
        else:
            duration[i]='0h '+duration[i]
```

```python
# die spalte Duration umwandeln
train_data['Duration']=duration
```

```python
train_data.tail()
```

```python
from pytimeparse.timeparse import timeparse # um die zahl in sekunde zu rechnen!
```

```python
'2h 35m'.split(' ')[1][0:-1]
```

```python
# 2/
def hour(x):
    return x.split(' ')[0][0:-1]

def minute(x):
    return x.split(' ')[1][0:-1]
```

```python
# die stunde und minute trennen
train_data['Duration_hours']=train_data['Duration'].apply(hour)
train_data['Duration_mins']=train_data['Duration'].apply(minute)
```

```python
train_data.head()
```

```python
# die spalte Duration löschen , nicht mehr gebrauchbar ----
drop_column(train_data,'Duration')
```

```python
train_data.head()
```

```python
# die 'Duration_hours' und 'Duration_mins ' zu int umwandeln
train_data['Duration_hours']=train_data['Duration_hours'].astype(int)
train_data['Duration_mins']=train_data['Duration_mins'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import metrics
```

```python
def predict(ml_model):
    model = ml_model.fit(X_train,y_train)
    print(f'Training score: {model.score(X_train,y_train)}')
    y_prediction=model.predict(X_test)
    print(f'predictions are:\n {y_prediction}')
    print('\n')
    
    r2_score=metrics.r2_score(y_test,y_prediction)
    print(f'r2 score is:{r2_score}')
    
    print('MAE: ', metrics.mean_absolute_error(y_test,y_prediction))
    print('MSE: ', metrics.mean_squared_error(y_test,y_prediction))
    print('RMSE: ',np.sqrt(metrics.mean_squared_error(y_test,y_prediction)))
    
    sns.displot(y_test-y_prediction)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg_rf = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=6)]
max_depth=[int(x) for x in np.linspace(start=5,stop=30,num=4)]
```

```python
RandomForestRegressor()
```

```python
random_grid={
    'n_estimators': n_estimators,
    'max_features':['auto','sqrt'],
    'max_depth':max_depth,
    'min_samples_split':[5,10,14,100]
}
```

```python
rf_random = RandomizedSearchCV(estimator=reg_rf,param_distributions=random_grid,cv=3,verbose=2,n_jobs=-1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
sns.distplot(y_test-prediction)
```

```python
metrics.r2_score(y_test,prediction)
```
