## Bild Einlesen


```python
from PIL import Image

image = Image.open('ein_simpson1.jpg')
image = image.resize((32,32),Image.LANCZOS) #jpg (3 farbe Kan√§len)
image
```

```python
import numpy as np
from keras.applications.vgg16 import preprocess_input

image = np.asarray(image).reshape(1,32,32,3)
image = preprocess_input(image)
```

```python
image.shape #(vgg16 --Y (1,32,32,3))
```

## VGG16 Laden und X_after_vgg berechnen

```python
from keras.applications.vgg16 import VGG16
vgg16_model = VGG16(include_top=False,input_shape=(32,32,3))
X_after_vgg = vgg16_model.predict(image)
```

```python
X_after_vgg.shape
```

## Model Laden und Wahrscheinlichkeit Berechnen

```python
from keras.models import load_model

model = load_model('model_vgg_simpson.h5')
model.predict(X_after_vgg)
```

## Wo Genau befindet sisch ein simpson

```python
filename = 'Simpson_Model.h5'

model_ = load_model(filename)
```

```python
img = Image.open('Main_simpson.jpg')
img = img.resize((round(img.size[0]/5),round(img.size[1]/5)),resample=Image.BICUBIC)
img
```

```python
print(np.asarray(img).shape,img.size)
```

```python
from PIL import ImageDraw
%matplotlib inline

#size = 100
sizes = [75,50,100,125]
step_size = 45

koordinaten = []

for size in sizes:
    for x in range(0,img.size[0]- size , step_size):
        for y in range(0,img.size[1]-size, step_size):
            part = img.crop((x,y, x + size , y + size))
            data = np.asarray(part.resize((32,32) , resample=Image.BICUBIC))
            data = data.astype(np.float32)/255


            pred = model_.predict(data.reshape(-1,32,32,3))
            #print(pred[0][0])
            if pred[0][0] > 0.98:
                koordinaten.append((x,y,size))
            
#print(simpsons_coordinaten)
out = img.copy()
draw = ImageDraw.Draw(out)

k_drawn_list = []

for k in koordinaten:
    exists = False
    for k_drawn in k_drawn_list:
        if k[0] >= k_drawn[0] and k[0] <= k_drawn[0] + k_drawn[2]: # + size
            if k[1] >= k_drawn[1] and k[1] <= k_drawn[1] + k_drawn[2]: # + size
                exists = True
    
    
    if exists == False:
        #points = [
        #    k,
        #    (k[0] + size , k[1] + size)
        #]
        #draw.rectangle(points)
        points = [
            (k[0],k[1]),
            (k[0],k[1] + k[2]),
            (k[0] + k[2] , k[1] + k[2]),
            (k[0] + k[2],k[1]),
            (k[0],k[1])
        ]
        draw.line(points,'yellow',3)
        k_drawn_list.append(k)
        print(k)
```
